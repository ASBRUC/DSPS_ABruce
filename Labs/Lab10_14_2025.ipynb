```python
import numpy as np
from matplotlib import pylab as plt
```

```python
np.random.seed(505)
X = (np.random.randn(100) * 3.5) + 12 #stretching the distribution
```

```python
X.mean()
```

```python
plt.hist(X, bins = 10) #Square root of points number is good idea
```

```python
X_minmax = X - X.min() #Shifted X-axis to 0
plt.hist(X_minmax, bins = 10)
```

```python
X_minmax = X_minmax / X_minmax.max()
plt.hist(X_minmax, bins = 10)
```

```python
X_standard = X - X.mean() #making mean = 0
plt.hist(X_standard, bins = 10)
```

```python
X_standard.mean()
```
