```python
import pandas as pd
```

```python
earthquakes = pd.read_csv("https://raw.githubusercontent.com/fedhere/DSPS_FBianco/refs/heads/main/data/earthquakes.csv",
            sep=r"\s+")
```

```python
earthquakes.shape #shows rows, cols
```

```python
earthquakes.head(); #show top (parameter) rows
earthquakes.tail() #show bottom (parameter) rows
```

```python
earthquakes.columns #shows all col names
```

```python
earthquakes = earthquakes[['#YYY/MM/DD', 'HH:mm:SS.ss','MAG']]
```

```python
earthquakes.dtypes #show data types of each variable
```

```python
from datetime import time
earthquakes.rename({"#YYY/MM/DD": "date","HH:mm:SS.ss": "time","MAG": "mag"},axis = 1, inplace=True) #{} is for dictionaries, axis = 1 to rename by col
```

```python
earthquakes.describe()
```

```python
earthquakes["newdates"] = earthquakes["date"].map(pd.to_datetime) #map applies the function to all elements in a series
```

```python
earthquakes["datetime"] = earthquakes["date"] + " " + earthquakes["time"]
earthquakes["datetime"]
```

```python
for i in earthquakes.index:
  if earthquakes.loc[i, "datetime"].endswith('60.00'):
    earthquakes.loc[i, "datetime"] = earthquakes.loc[i, "datetime"][:6] + "59,99"
```

```python
earthquakes["newdatetime"] = earthquakes["datetime"].map(pd.to_datetime)
```

```python
earthquakes["newdatetime"].std()
earthquakes["newdatetime"].mean()
earthquakes["newdatetime"].diff() #takes difference from last point to current point
```
