```python
import pandas as pd
```

```python
earthquakes = pd.read_csv("https://raw.githubusercontent.com/fedhere/DSPS_FBianco/refs/heads/main/data/earthquakes.csv",
            sep=r"\s+")
```

```python
earthquakes.shape #shows rows, cols
```

```python
earthquakes.head(); #show top (parameter) rows
earthquakes.tail() #show bottom (parameter) rows
```

```python
earthquakes.columns #shows all col names
```

```python
earthquakes = earthquakes[['#YYY/MM/DD', 'HH:mm:SS.ss','MAG']]
```

```python
earthquakes.dtypes #show data types of each variable
```

```python
from datetime import time
earthquakes.rename({"#YYY/MM/DD": "date","HH:mm:SS.ss": "time","MAG": "mag"},axis = 1, inplace=True) #{} is for dictionaries, axis = 1 to rename by col
```
